```python
# UNZIP DataTeamShared Data
def un_gz(file_name, dirs):
    t = tarfile.open(file_name)
    dirs1 = dirs + '/' + os.path.basename(file_name)
    os.mkdir(dirs1)
    t.extractall(path=dirs1)
    da_te = os.path.basename(dirs)
    readPath = dirs1 + '/Logs_full/md***' + da_te + '***'
    dataPathLs = np.array(glob.glob(readPath))
    assert(len(dataPathLs) > 0)
    readPath = dirs1 + '/Logs_full/***'
    dataPathLs1 = np.array(glob.glob(readPath))
    for i in list(set(dataPathLs1) - set(dataPathLs)):
        if os.path.exists(i):
            if os.path.isfile(i):
                os.remove(i)
            if os.path.isdir(i):
                shutil.rmtree(i) 
    print('finish ' + file_name)

import tarfile
import os
import numpy as np
import glob
import shutil

try:
    os.mkdir('/mnt/data/' + str(date))
except:
    print('folder already created')

for path in np.sort(dataPathLs):
    un_gz(path, '/mnt/data/' + str(date))
```

```python
# UNZIP SH/SZ api data
import tarfile
import os
import numpy as np
import glob
import shutil

date = 20210112
exchange = 'SZ'

def un_gz1(file_name, dirs):
    t = tarfile.open(file_name)
    dirs1 = dirs + '/' + os.path.basename(file_name)
    os.mkdir(dirs1)
    t.extractall(path=dirs1)
    da_te = os.path.basename(dirs)
    readPath = dirs1 + '/md***' + da_te + '***'
    dataPathLs = np.array(glob.glob(readPath))
    assert(len(dataPathLs) > 0)
    readPath = dirs1 + '/***'
    dataPathLs1 = np.array(glob.glob(readPath))
    for i in list(set(dataPathLs1) - set(dataPathLs)):
        if os.path.exists(i):
            if os.path.isfile(i):
                os.remove(i)
            if os.path.isdir(i):
                shutil.rmtree(i) 
    print('finish ' + file_name)

readPath1 = '/mnt/dailyRawData/' + str(date) + '/***zs_96_03_day_96data.tar.gz'
# readPath1 = '/mnt/dailyRawData/' + str(date) + '/***zs_92_01_day_data.tar.gz'
readPath2 = '/mnt/dailyRawData/' + str(date) + '/***zt_88_03_day_88data.tar.gz'
# readPath2 = '/mnt/dailyRawData/' + str(date) + '/***zt_96_04_day_96data.tar.gz'
dataPathLs1 = np.array(glob.glob(readPath1))
assert(len(dataPathLs1) == 1)
dataPathLs2 = np.array(glob.glob(readPath2))
assert(len(dataPathLs2) == 1)

if exchange == 'SH':
    dataPathLs = dataPathLs2
elif exchange == 'SZ':
    dataPathLs = dataPathLs1
else:
    dataPathLs = list(dataPathLs1) + list(dataPathLs2)
print(dataPathLs)

try:
    os.mkdir('/mnt/data/' + str(date))
except:
    print('folder already created')
    
for path in np.sort(dataPathLs):
    un_gz1(path, '/mnt/data/' + str(date))
```
